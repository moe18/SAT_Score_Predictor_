```python
import pandas 
import numpy
import seaborn
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
SAT=pandas.read_excel('2012SAT scores_02.xls')
survey=pandas.read_excel('HS Survey Results_02.xlsx')
```

```python
SAT.head()
sheet = pandas.merge(SAT, survey, on='DBN')
SAT['Num of SAT Test Takers']
```

```python
Enrollment=sm.OLS(sheet['SAT Total'],
              sheet['Enrollment']).fit() 

Enrollment.summary()
```

```python
%matplotlib inline
seaborn.regplot(y='SAT Total', x='Enrollment', data=sheet, fit_reg=True)  
plt.ylabel("SAT score")
plt.xlabel('number enrolled')
plt.title("enrollment v.s SAT score")
```
